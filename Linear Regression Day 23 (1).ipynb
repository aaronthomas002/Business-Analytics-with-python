```python
import pandas as pd
```

```python
df=pd.read_csv("Sales_data.csv")
```

```python
df
```

```python
x=df.iloc[:, :-1].values
```

```python
x
```

```python
y=df.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
y_pred=regressor.predict(x)
```

```python
y_pred
```
