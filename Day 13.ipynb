```python
import matplotlib.pyplot as plt
```

```python
x=[12,15,16,18,19]
y=[13,14,18,19,22]

```

```python
plt.plot(x,y)
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,marker='o')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'--')
plt.ylabel('sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'*')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'3')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'r')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'b')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'o:r')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'o:r',ms=20, mec ='b')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'o:r',ms=20, mec ='r',mfc='g')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[20,25,30,35,40]
z=[45,50,55,60,65]
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'--',ms=20)
plt.plot(x,z,'o:r',ms=20,mec='r',mfc='b')
plt.ylabel('Sales')
plt.xlabel('Revenue')
plt.show()
```

```python
plt.plot(x,y,'--',ms=20)
plt.plot(x,z,'o:r',ms=20,mec='r',mfc='b')
plt.ylabel('Sales',fontdict = {'color':'blue','size':20})
plt.xlabel('Revenue',fontdict = {'color':'blue','size':20})
plt.show()
```

```python
plt.plot(x,y,'--',ms=20)
plt.plot(x,z,'o:r',ms=20,mec='r',mfc='b')
plt.ylabel('Sales',fontdict = {'family':'serif','color':'blue','size':20})
plt.xlabel('Revenue',fontdict = {'family':'serif','color':'green','size':20})
plt.show()
```

```python
plt.plot(x,y,'--',ms=20)
plt.plot(x,z,'o:r',ms=20,mec='r',mfc='b')
plt.ylabel('Sales',fontdict = {'family':'serif','color':'blue','size':20})
plt.xlabel('Revenue',fontdict = {'family':'serif','color':'green','size':20})
plt.title(" HR Graph")
plt.show()
```
