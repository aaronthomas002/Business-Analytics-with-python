```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([7,14,24,30,45])
```

```python
## ynew(line)=mx+c
## m=sigma(x-xdash)(y-ydash)/sigma(x-xdash)square
## c=ydash-m*xdash
```

```python
xdash=x.mean()
```

```python
xdash
```

```python
ydash=y.mean()
ydash
```

```python
A=x-xdash
print(A)
```

```python
B=y-ydash
print(B)
```

```python
C=sum(A*B)
print(C)
```

```python
D=sum(A*A)
print(D)
```

```python
m=C/D
m
```

```python
c=ydash-m*xdash
c
```

```python
ynew=m*x+c
ynew
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,ynew,)
plt.scatter(x,y)
plt.xlabel('days')
plt.ylabel('Sales')
```
